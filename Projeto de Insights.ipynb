### EXERCÍCIOS PT2

```python
#Convertendo a coluna 'date' em datetime

df['date'] = pd.to_datetime(df['date'], format="%Y-%m-%d")
```

```python
#Mostrando as 5 primeiras linhas do DF para conferência da coluna Date

df.head()
```

```python
#1: Crie uma coluna house_age que indique se a casa é nova ou velha a partir da data de venda

df['house_age'] = None
df.loc[df.date >= '2014-01-01', 'house_age'] = 'new_house'
df.loc[df.date < '2014-01-01', 'house_age'] = 'old_house'
```

```python
df['house_age'].value_counts()
```

```python
#2: Crie uma coluna dormitory_types que classifique os imóveis a partir do número de quartos
#Imóveis com 1 quarto: Studio;
#Imóveis com 2 quartos: Apartment;
#Imóveis com mais de 2 quartos: house.

df['dormitory_types'] = None
df.loc[df.condition == 1, 'dormitory_types'] = 'Studio'
df.loc[df.bedrooms == 2, 'dormitory_types'] = 'Apartment'
df.loc[df.bedrooms > 2, 'dormitory_types'] = 'house'
```

```python
df['dormitory_types'].value_counts()
```

```python
df.head()
```

```python
#3: Crie a coluna condition_type para classificar os imóveis de acordo com sua condição.
#Imóveis com condição menor ou igual a 2: Bad;
#Imóveis com condição entre 3 e 4: Regular;
#Imóveis com condição 5: Good.

df['condition_type'] = None
df.loc[df.condition <= 2, 'condition_type'] = 'Bad'
df.loc[df.condition ==3, 'condition_type'] = 'Regular'
df.loc[df.condition ==4, 'condition_type'] = 'Regular'
df.loc[df.condition >=5, 'condition_type'] = 'Good'
```

```python
df['condition_type'].value_counts()
```

```python
#Questão 4: Converta a coluna condition para str

df.condition=df['condition'].astype(str)
```

```python
df.dtypes
```

```python
#5: Apague as colunas sqft_living15 e sqft_lot15


df=df.drop('sqft_living15',1)
df=df.drop('sqft_lot15',1)
df.columns
```

```python
#6: Converta a coluna yr_built para datetime

df['yr_built'] = pd.to_datetime(df['yr_built'], format="%Y-%m-%d")
df.head()
```

```python
#7: Converta a coluna yr_renovated para datetime

df['yr_renovated'] = pd.to_datetime(df['yr_renovated'], format="%Y", errors='coerce') #é necessário utilizar o parâmetro errors='coerce' que faz com que os valores que não consigam fazer a conversão se torne NaT (Not a Time)
df.head()
```

```python
#8. Qual a data mais antiga de construção de um imóvel?
#Resposta: A data mais antida de construção de um imóvel é 01/01/1900.


df.yr_built.min()
```

```python
#9. Qual a data mais antiga de renovação de um imóvel?
#Resposta: A data mais antida de renovação de um imóvel é 01/01/1934.

df.yr_renovated.min()
```
